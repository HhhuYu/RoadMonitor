```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
file_path = '../../data/Final_Version/poor/590/ACCL/new3/d37e6_acc_590_1.csv'
data = pd.read_csv(file_path)
```

```python
x = data['x']
y = data['y']
z = data['z']

x2 = data['x2']
y2 = data['y2']
z2 = data['z2']

xFilter = data['xFilter']
yFilter = data['yFilter']
zFilter = data['zFilter']
t = data['time']
```

```python
def timeToRowNum(time):
    return int(time*50)
```

```python
def plotWaveform(st_time, ed_time, ticker_size):
    rowNum1, rowNum2 = timeToRowNum(st_time), timeToRowNum(ed_time)
    plt.figure(figsize=(25, 10))

    # plot x
    plt.subplot(331)
    plt.grid(True)
    plt.plot((t[rowNum1:rowNum2]), x[rowNum1:rowNum2], label='AccX')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    plt.ylabel('Amplitude',size=14)
    plt.ylim([-5, 5])
    plt.title('Raw data',size=16)
    # plot x_
    plt.subplot(332)
    plt.grid(True)
    plt.plot((t[rowNum1:rowNum2]), x2[rowNum1:rowNum2], label='AccX_reoriented')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    #plt.ylabel('Acc(m/s^2)')
    plt.ylim([-5, 5])
    plt.title('Reoriented data',size=16)
    # plot x_Filter
    plt.subplot(333)
    plt.grid(True)
    plt.plot((t[rowNum1:rowNum2]),
             xFilter[rowNum1:rowNum2], label='AccX_filtered')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    #plt.ylabel('Acc(m/s^2)')
    plt.ylim([-5, 5])
    plt.title('Filtered data',size=16)

    # plot y
    plt.subplot(334)
    plt.grid(True)
    plt.plot((t[rowNum1:rowNum2]), y[rowNum1:rowNum2], label='AccY')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    plt.ylabel('Amplitude',size=14)
    plt.ylim([-5, 5])
    # plot y_
    plt.subplot(335)
    plt.grid(True)
    plt.plot((t[rowNum1:rowNum2]), y2[rowNum1:rowNum2], label='AccY_reoriented')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    #plt.ylabel('Acc(m/s^2)')
    plt.ylim([-5, 5])
    # plot y_f
    plt.subplot(336)
    plt.grid(True)
    plt.plot((t[rowNum1:rowNum2]),
             yFilter[rowNum1:rowNum2], label='AccY_filtered')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    #plt.ylabel('Acc(m/s^2)')
    plt.ylim([-5, 5])

    # plot z
    plt.subplot(337)
    plt.grid(True)
    plt.plot((t[rowNum1:rowNum2]), z[rowNum1:rowNum2], label='AccZ')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    plt.ylabel('Amplitude',size=14)
    plt.xlabel('Time (s)',size=14)
    plt.ylim([5, 15])
    # plot z_
    plt.subplot(338)
    plt.grid(True)
    plt.plot((t[rowNum1:rowNum2]), z2[rowNum1:rowNum2], label='AccZ_reoriented')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    #plt.ylabel('Acc(m/s^2')
    plt.xlabel('Time (s)',size=14)
    plt.ylim([5, 15])
    # plot z_f
    plt.subplot(339)
    plt.grid(True)
    plt.plot((t[rowNum1:rowNum2]),
             zFilter[rowNum1:rowNum2], label='AccZ_filtered')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    #plt.ylabel('Acc(m/s^2')
    plt.xlabel('Time (s)',size=14)
    plt.ylim([-5, 5])

    plt.savefig('raw-reoriented-filtered.svg',bbox_inches='tight',pad_inches = 0)
```

```python
a = 40
plotWaveform(a-5, a+5, 5)
```
