```python
import math
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pywt
%matplotlib inline
```

```python
data_file = '../../data/Final_Version/poor/590/ACCL/new3/d37e6_acc_590_1.csv'
data = pd.read_csv(data_file, encoding='utf-8')
print(data.shape)
```

```python
x = data['xFilter']
y = data['yFilter']
z = data['zFilter']

t = data['time']
```

```python
def timeToRowNum(time):
    return int(time*50)
```

```python
def plotWaveform(st_time0, ed_time0, st_time1, ed_time1,st_time2, ed_time2, ticker_size):
    rowNum,rowNum0 =  timeToRowNum(st_time0), timeToRowNum(ed_time0)
    rowNum1, rowNum2 = timeToRowNum(st_time1), timeToRowNum(ed_time1)
    rowNum3, rowNum4 = timeToRowNum(st_time2), timeToRowNum(ed_time2)
    plt.figure(figsize=(25, 7))
    
    # plot x0
    plt.subplot(231)
    plt.grid(True)
    plt.plot((t[rowNum:rowNum0]), x[rowNum:rowNum0], label='AccX_filtered')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    plt.ylabel('Amplitude',size=14)
    plt.ylim([-5, 5])
    plt.title('Normal',size=16)
    # plot x
    plt.subplot(232)
    plt.grid(True)
    plt.plot((t[rowNum1:rowNum2]), x[rowNum1:rowNum2], label='AccX_filtered')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    #plt.ylabel('Acc(m/s^2)')
    plt.ylim([-5, 5])
    plt.title('Transverse',size=16)
    # plot x_
    plt.subplot(233)
    plt.grid(True)
    plt.plot((t[rowNum3:rowNum4]), x[rowNum3:rowNum4], label='AccX_filtered')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    #plt.ylabel('Acc(m/s^2)')
    plt.ylim([-5, 5])
    plt.title('Pothole',size=16)
    
    # plot z0
    plt.subplot(234)
    plt.grid(True)
    plt.plot((t[rowNum:rowNum0]), z[rowNum:rowNum0], label='AccZ_filtered')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    plt.ylabel('Amplitude',size=14)
    plt.xlabel('Time (s)',size=14)
    plt.ylim([-5, 5])
    
    # plot z
    plt.subplot(235)
    plt.grid(True)
    plt.plot((t[rowNum1:rowNum2]), z[rowNum1:rowNum2], label='AccZ_filtered')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    #plt.ylabel('Acc(m/s^2)')
    plt.xlabel('Time (s)',size=14)
    plt.ylim([-5, 5])
    # plot z_
    plt.subplot(236)
    plt.grid(True)
    plt.plot((t[rowNum3:rowNum4]), z[rowNum3:rowNum4], label='AccZ_filtered')
    plt.legend(loc=1)
    ax = plt.gca()
    ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(
        ticker_size))  # Refine the axis, interval= ticker_size
    #plt.ylabel('Acc(m/s^2')
    plt.xlabel('Time (s)',size=14)
    plt.ylim([-5, 5])
   
    plt.savefig('x-zRation.svg',bbox_inches='tight',pad_inches = 0)
```
